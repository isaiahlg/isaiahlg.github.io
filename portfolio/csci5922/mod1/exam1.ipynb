For y and y^ as close, let's use y^ = 0.9 and y = 1 and calculate the loss:
```
L = -[1 * log(0.9) + (1-1) * log(1-0.9)]  
  = -[1 * log(0.9) + 0 * log(1-0.9)]  
  = -[1 * log(0.9) + 0 * log(0.1)]  
  = -[1 * -0.1054 + 0 * -2.3026]  
  = -[-0.1054 + 0]  
  = 0.1054  
```

For y and y^ as far, let's use y^ = 0.1 and y = 1 and calculate the loss:
```
L = -[1 * log(0.1) + (1-1) * log(1-0.1)]  
  = -[1 * log(0.1) + 0 * log(1-0.1)]  
  = -[1 * log(0.1) + 0 * log(0.9)]  
  = -[1 * -2.3026 + 0 * -0.1054]  
  = -[-2.3026 + 0]  
  = 2.3026  
```
We can see that the loss is much greater (2.306) when y and y^ are far apart than when they are close (0.1054).
